```python
import numpy as np
```

```python
def quadratic_time_algorithm(A, B, n):

    out = np.zeros(2*n-1)

    for i in range(n):
        for j in range(n):
            out[i+j] += A[i]* B[j]
    
    return out

def karatsuba_algorithm(A, B, n):
    
    if n == 1:
        return [A[0] * B[0]]
    
    size = int(n//2)
    
    p_h = A[:size]
    p_l = A[size:]

    q_h = B[:size]
    q_l = B[size:]

    r_x = karatsuba_algorithm(p_h, q_h, size)
    s_x = karatsuba_algorithm(p_l, q_l, size)
    t_x = karatsuba_algorithm(([p_h[i] + p_l[i] for i in range(size)]), ([q_h[i] + q_l[i] for i in range(size)]), size)

    r_3 = [t_x[i] - r_x[i] - s_x[i] for i in range(len(t_x))]
                   
    out = np.zeros(2*n-1)

    for i, val in enumerate(r_x):
        out[i] += val
    
    for i, val in enumerate(r_3):
        out[size + i] += val
    
    for i, val in enumerate(s_x):
        out[2 * size + i] += val

    return out


def hybrid_algorithm(A, B, n, T):
  if (n <= T):
    return quadratic_time_algorithm(A, B, n)
  else:
    return karatsuba_algorithm(A, B, n)
```
